```python
# Install necessary libraries
!pip install nltk
!pip install transformers

# Import libraries
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer
from nltk.tokenize import word_tokenize
import torch
from transformers import AutoModelForSequenceClassification, AutoTokenizer

# Download NLTK data
nltk.download('vader_lexicon')
nltk.download('stopwords')
nltk.download('punkt')

# Load pre-trained sentiment analysis model
model_name = "distilbert-base-uncased-finetuned-sst-2-english"
model = AutoModelForSequenceClassification.from_pretrained(model_name)
tokenizer = AutoTokenizer.from_pretrained(model_name)

# Define chatbot function
def chatbot(user_input):
    # Preprocess user input
    inputs = tokenizer(user_input, return_tensors='pt')

    # Analyze sentiment
    with torch.no_grad():
        logits = model(**inputs).logits
        probabilities = torch.nn.functional.softmax(logits, dim=1)
        positive_prob = probabilities[:, 1].item()
        negative_prob = probabilities[:, 0].item()

    # Generate response based on sentiment
    threshold = 0.6  # adjust this value as needed
    if positive_prob > threshold:
        return "You seem happy! What's making you smile?"
    elif negative_prob > threshold:
        return "Sorry to hear that. Would you like to talk about it?"
    else:
        return "That's interesting! Can you tell me more?"

# Test chatbot
user_input = input("")
print(chatbot(user_input))

user_input =input("")
print(chatbot(user_input))

user_input = input("")
print(chatbot(user_input))

```
